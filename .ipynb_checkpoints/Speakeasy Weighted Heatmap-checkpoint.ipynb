```python
import matplotlib.pyplot as plt
from matplotlib import colors as mcolors
import pandas as pd
import gmaps
import datetime
import config

# Configure gmaps
gmaps.configure(api_key=config.api_key)

file = "speakeasies.csv"
df = pd.read_csv(file)
df.head()
```

```python
#set df to index by rating
reindexed_data = df.set_index('Rating')
reindexed_data.head()
```

```python
fig = gmaps.figure(map_type="HYBRID")
heatmap_layer = gmaps.heatmap_layer(
        reindexed_data[['Lat','Lng']].values,
        weights=reindexed_data.index.values
    )
fig.add_layer(heatmap_layer)
fig
```
